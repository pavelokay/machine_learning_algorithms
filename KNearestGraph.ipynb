```python
import json
import matplotlib.pyplot as plt
import os
data = {}
data['points'] = []
rangeX_Y = [50 , 250]
# SET PATH TO FILE WITH POINTS.
with open('unlabeledPointsKNearest.json') as json_file:
    data = json.load(json_file)
    for p in data:
        point = p['Parameters']
        plt.scatter(point[0], point[1], s=150, c='black')
    # Set chart title.
    plt.title("Unlabeled Points - TestData", fontsize=19)

    # Set x axis label.
    plt.xlabel("x", fontsize=10)

    # Set y axis label.
    plt.ylabel("y", fontsize=10)
    # Set size of tick labels.
    plt.tick_params(axis='both', which='major', labelsize=9)
    plt.xlim([0 , 100])
    plt.ylim(rangeX_Y)
    plt.show()
with open('trainingKNearest.json') as json_file:
    data = json.load(json_file)
    for p in data:
        point = p['Parameters']
        classification = p['Classification']
        if classification == 0:
            color = 'red'
        if classification == 1:
            color = 'blue'
        if classification == 2:
            color = 'green'
        if classification == 3:
            color = 'yellow'
        if classification == 4:
            color = 'brown'
        plt.scatter(point[0], point[1], s=50, c=color)
    # Set chart title.
    plt.title("Training Points", fontsize=19)

    # Set x axis label.
    plt.xlabel("x", fontsize=10)

    # Set y axis label.
    plt.ylabel("y", fontsize=10)
    # Set size of tick labels.
    plt.tick_params(axis='both', which='major', labelsize=9)
    plt.xlim([0 , 100])
    plt.ylim(rangeX_Y)
    plt.show()
with open('trainingKNearest.json') as json_file:
    data = json.load(json_file)
    for p in data:
        point = p['Parameters']
        classification = p['Classification']
        if classification == 0:
            color = 'red'
        if classification == 1:
            color = 'blue'
        if classification == 2:
            color = 'green'
        if classification == 3:
            color = 'yellow'
        if classification == 4:
            color = 'brown'
        plt.scatter(point[0], point[1], s=50, c=color)   
    with open('labaledPointsKNearest.json') as json_fileC:
        dataC = json.load(json_fileC)
        for f in dataC:
            pointC = f['Parameters']
            classificationC = f['Classification']
            if classificationC == 0:
                colorC = 'red'
            if classificationC == 1:
                colorC = 'blue'
            if classificationC == 2:
                colorC = 'green'
            if classificationC == 3:
                colorC = 'yellow'
            if classificationC == 4:
                colorC = 'brown'
            plt.scatter(pointC[0], pointC[1], s=250, c=colorC)
    # Set chart title.
    plt.title("Labeled Points", fontsize=19)

    # Set x axis label.
    plt.xlabel("x", fontsize=10)

    # Set y axis label.
    plt.ylabel("y", fontsize=10)
    # Set size of tick labels.
    plt.tick_params(axis='both', which='major', labelsize=9)
    plt.xlim([0 , 100])
    plt.ylim(rangeX_Y)
    plt.show()

            
        
        



```
